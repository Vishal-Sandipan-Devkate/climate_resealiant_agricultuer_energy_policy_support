```python
def fill_energy_by_year(row):
    y = row['year']

    # ---- SOLAR MW ----
    if pd.isna(row['solar_mw_district']):
        if y <= 2018:
            row['solar_mw_district'] = 0.01
        elif 2019 <= y <= 2021:
            row['solar_mw_district'] = 0.05
        elif 2022 <= y <= 2023:
            row['solar_mw_district'] = 0.10
        else:  # 2024â€“2025
            row['solar_mw_district'] = 0.20

    # ---- WIND (always zero) ----
    if pd.isna(row['wind_mw']):
        row['wind_mw'] = 0.0

    # ---- BIOENERGY ----
    if pd.isna(row['bioenergy_mw']):
        if y <= 2020:
            row['bioenergy_mw'] = 0.0
        else:
            row['bioenergy_mw'] = 0.01

    # ---- SOLAR PUMPS ----
    if pd.isna(row['solar_pumps_installed']):
        if y <= 2018:
            row['solar_pumps_installed'] = 5
        elif 2019 <= y <= 2021:
            row['solar_pumps_installed'] = 25
        elif 2022 <= y <= 2023:
            row['solar_pumps_installed'] = 60
        else:
            row['solar_pumps_installed'] = 120

    # ---- CO2 REDUCTION (derived) ----
    if pd.isna(row['estimated_co2_reduction_tonnes']):
        row['estimated_co2_reduction_tonnes'] = (
            row['solar_mw_district'] * 1200
        )

    return row
```

```python
df = df.apply(fill_energy_by_year, axis=1)
```
