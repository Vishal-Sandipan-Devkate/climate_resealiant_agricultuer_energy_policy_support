```python
import pandas as pd

# Load district master
districts = pd.read_csv("D:/imp data sets/india_2011_district.csv")

districts
```

```python
#Standardize
districts['st_nm'] = districts['st_nm'].str.lower().str.strip()
districts['district'] = districts['district'].str.lower().str.strip()

# Add time dimension
years = range(2015, 2026)
months = range(1, 13)

base = (
    districts[['st_nm', 'district']]
    .assign(key=1)
    .merge(
        pd.DataFrame([(y, m) for y in years for m in months],
                     columns=['year', 'month']).assign(key=1),
        on='key'
    )
    .drop('key', axis=1)
)

print("Base rows:", len(base))
base.head()
```

```python
solar_state.columns = solar_state.columns.str.strip()
```

```python
solar_state = solar_state.rename(columns={
    'Solar Power (MWp)': 'solar_mw',
    'Wind Power (MW)': 'wind_mw',
    'Bio-Power-BM Power/Cogen.(MW)': 'bioenergy_bm_mw',
    'Bio-Power-Waste to Energy (MW)': 'bioenergy_wte_mw'
})
```

```python
solar_state['bioenergy_mw'] = (
    solar_state['bioenergy_bm_mw'].fillna(0) +
    solar_state['bioenergy_wte_mw'].fillna(0)
)
```

```python
# Keep only one n_districts column
solar_state['n_districts'] = solar_state['n_districts_x']

solar_state = solar_state.drop(
    columns=['n_districts_x', 'n_districts_y'],
    errors='ignore'
)
```

```python
solar_state['solar_mw_district'] = solar_state['solar_mw'] / solar_state['n_districts']
solar_state['wind_mw_district']  = solar_state['wind_mw'] / solar_state['n_districts']
solar_state['bioenergy_mw_district'] = solar_state['bioenergy_mw'] / solar_state['n_districts']
```

```python
base = base.merge(
    solar_state[[
        'state',
        'solar_mw_district',
        'wind_mw_district',
        'bioenergy_mw_district'
    ]],
    on='state',
    how='left'
)
```

```python
agri = pd.read_csv("D:/imp data sets/RS_Session_266_AU_1739_C_2.csv")

agri.head()
```

```python
agri.shape
```

```python
print(agri.columns.tolist())
```

```python
agri.columns = agri.columns.str.lower().str.strip()
```

```python
print(agri.columns.tolist())
```

```python
agri = agri.rename(columns={
    'state/ut': 'state',
    'component-a - installed (mw)': 'solar_agri_mw',
    'component-b - installed (in numbers)': 'solar_pumps_installed',
    'component-c - installed (total) (in numbers)': 'solarized_grid_pumps'
})
```

```python
agri['state'] = agri['state'].str.lower().str.strip()
```

```python
district_counts = base.groupby('state')['district'].nunique().reset_index(name='n_districts')

agri = agri.merge(district_counts, on='state', how='left')

agri['solar_pumps_district'] = agri['solar_pumps_installed'] / agri['n_districts']

base = base.merge(
    agri[['state', 'solar_pumps_district']],
    on='state',
    how='left'
)
```

```python
base = base.rename(columns={'solar_pumps_district': 'solar_pumps_installed'})
```

```python
GRID_EF = 0.82  # kg CO2 per kWh

base['solar_gen_mwh'] = base['solar_mw_district'] * 0.18 * 24 * 30
base['estimated_co2_reduction_tonnes'] = base['solar_gen_mwh'] * GRID_EF
```

```python
print([c for c in base.columns if 'solar' in c.lower()])
```

```python
# keep one solar column
base['solar_mw_district'] = base['solar_mw_district_x']

# drop duplicates
base = base.drop(
    columns=['solar_mw_district_x', 'solar_mw_district_y'],
    errors='ignore'
)
```

```python
print([c for c in base.columns if 'solar' in c.lower()])
```

```python
# constants
SOLAR_CF = 0.18        # capacity factor
HOURS_PER_MONTH = 24 * 30

base['solar_gen_mwh'] = (
    base['solar_mw_district'] * SOLAR_CF * HOURS_PER_MONTH
)
```

```python
GRID_EF = 0.82  # kg CO2 per kWh

base['estimated_co2_reduction_tonnes'] = (
    base['solar_gen_mwh'] * GRID_EF
)
```
